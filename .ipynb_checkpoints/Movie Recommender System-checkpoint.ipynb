```python
import pandas as pd
import numpy as np
```

```python
movies= pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
 # merge two datasets

movies=movies.merge(credits,on="title") 
movies.head(1)
```

```python
# important factors :-geners,id,keyword,title,overview,cast,crew

movies=movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.head(1)
```

```python
# Check null values

movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# Check duplicate values

movies.duplicated().sum()
```

# Data preprocessing  (creating new data frame having three columns        movie_id, title, tags)

```python
movies.iloc[0].genres
```

```python
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

#convert str into readable list format:
#                 ['Action','adventure','Fantacy','Scifi']
```

```python
import ast
def convert(obj):
    l1=[]
    for i in ast.literal_eval(obj):   # ast convert str value into list
        l1.append(i['name'])
    return l1
```

```python
movies['genres']=movies["genres"].apply(convert)
```

```python
movies['keywords']=movies["keywords"].apply(convert)
```

```python
movies.head(1)
```

```python
import ast
def convert3(obj):
    l1=[]
    counter=0
    for i in ast.literal_eval(obj):  
        if counter!=3:
            l1.append(i['name'])
        else:
            break
    return l1
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
import ast
def fetch_director(obj):
    l1=[]
    for i in ast.literal_eval(obj):  
        if i["job"]=="Director":
            l1.append(i['name'])
            break
    return l1
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
# Overview is str ,converting into list

movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Remove space between words

movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#concatenates 

movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
new_df=movies[["movie_id","title","tags"]]
```

```python
new_df.head()
```

```python
# List into string

new_df["tags"]=new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()                              #returns one word for similar words
```

```python
ps.stem("loved")
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
new_df["tags"][0]
```

# Vectorization (Convert text data to numerical vectors)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()  # By defoult cv returns scipi sparse matrix ,so convert into numpy array
```

```python
vectors[0]                    # tags converted into vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]  # similarity of first movie with all other 4806 movies
```

```python
sorted(list(enumerate(similarity[4])),reverse=True,key=lambda x:x[1]) [1:11]
```

```python
def recommend(movie):
    
    #sel_movie_id=new_df[new_df["title"]==movie].movie_id.values
   
    
    movie_index=new_df[new_df["title"]==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1]) [1:6]
    #print(sel_movie_id)
    #print(movie_list)
    #print(''.join(sel_movie_id))
    #print(movie_index)
    #print(type(sel_movie_id))
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
   
```

```python
recommend("My Date with Drew")
```
